## Are the variables "Smoker" and "Low Birthweight" correlated?
We start with a crosstabulation, then use a &chi;<sup>2</sup> test

```python
cx = pandas.crosstab(dataset.smoker,dataset.lowbwt)
cx
```

```python
stats.chi2_contingency(cx)
```

## Do smokers have smaller babies?
We look at the means for three variables for smokers and non-smokers. We then compare these means with a T-test

```python
non_smokers = dataset[dataset['smoker'] == 0]
smokers = dataset[dataset['smoker'] == 1]
cols = ['Length','Birthweight','Gestation']
print('Smokers:\n',smokers[cols].describe(),'\n\nNon-smokers:\n',non_smokers[cols].describe())
```

```python
for col in cols:
    print('\n', col, ':\n  ', stats.ttest_ind(smokers[col], non_smokers[col]))
```
